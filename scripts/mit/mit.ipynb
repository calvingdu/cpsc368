## Phase 3 – Data Collection from MIT datasets

This notebook collects MIT data and refines into CSV files that can then be uploaded to SQL Oracle


```python
import pandas as pd
import os
```

```python
df_1976_2020 = pd.read_csv('../../source_data/mit/1976-2020-president.csv')
df_2024 = pd.read_csv('../../source_data/mit/2024-president-state.csv')

df_2016 = df_1976_2020[df_1976_2020['year'] == 2016].reset_index(drop=True)
df_2020 = df_1976_2020[df_1976_2020['year'] == 2020].reset_index(drop=True)

print("2016:")
print(df_2016.shape[0], "rows")
display(df_2016.head())
print("2020:")
print(df_2020.shape[0], "rows")
display(df_2020.head())
print("2024:")
print(df_2024.shape[0], "rows")
display(df_2024.head())
```

```python
common_cols = df_2016.columns.intersection(df_2020.columns).intersection(df_2024.columns)
print('Common columns:', list(common_cols))

df = pd.concat([df_2016, df_2020, df_2024], ignore_index=True)[common_cols]
print(df.shape[0], "rows total amongst 2016, 2020, and 2024 datasets")
display(df.head())
display(df['year'].unique())

# Removing all parties except DEM and REP, and all rows with missing values in the 'candidate' column
df = df[df['party_simplified'].isin(['DEMOCRAT', 'REPUBLICAN'])]
df = df.dropna(subset=['candidate'])
print(df.shape[0], "rows after filtering for DEM and REP parties and dropping rows with missing candidate names")
display(df.head())
```

### Data Quality Testing

```python
assert(df['office'].nunique() == 1 and df['office'].unique()[0] == 'US PRESIDENT')
assert(df['state'].nunique() == 51)
```

### Upload

```python
os.makedirs('../../cleaned_data/mit', exist_ok=True)    
display(df.head())
df.rename(columns={'party_simplified': 'party', 'totalvotes': 'votes'}, inplace=True)
final_df = df[['year', 'state', 'state_po', 'candidate', 'party', 'votes']]
display(final_df.head())
final_df.to_csv('../../cleaned_data/mit/mit_presidential_election_data.csv', index=False)
```
