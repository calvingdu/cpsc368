## Phase 4 – Clean and upload ANES to MongoDB 
This notebook creates JSON files from cleaned ANES csv files and uploads it to MongoDB.

```python
import csv
import json
import pandas as pd
from pymongo import MongoClient
```

```python
# 1. create JSON of ANES data
anes2024 = pd.read_csv('anes_2024_clean_sample.csv')
anes2020 = pd.read_csv('anes_2020_clean_sample.csv')

# Add year column 
anes2020["election_year"] = 2020
anes2024["election_year"] = 2024

df = pd.concat([anes2020, anes2024], ignore_index=True)
df = df.astype(object).where(pd.notna(df), None)

# Drop ID columns
df = df.drop(columns=["id_2024", "id_2020", "id_2016"], errors="ignore")
documents = df.to_dict(orient="records")

# Save to JSON
with open("anes.json", "w") as f:
    json.dump(documents, f, indent=2)
```

```python
# NOTE: Ensure MongoDB is installed and running locally on port 27017
client = MongoClient("mongodb://localhost:27017/")
db = client["youtube_db"]
collection = db["channels"]

# Clear old data
collection.delete_many({})

# Insert documents
if documents:
    collection.insert_many(documents)

print("Data inserted into MongoDB")
```
